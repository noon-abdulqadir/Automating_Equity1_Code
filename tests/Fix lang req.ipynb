```python
import pandas as pd
import itertools
```

```python
df_save_dir = '/Users/nyxinsane/Documents/Work - UvA/Automating Equity/Study 1/Study1_Code/data/final dfs/'
```

```python
cols_list_df = pd.read_pickle(f'{df_save_dir}df_jobs_for_classification.pkl')
cols_list_ = cols_list_df.columns.tolist()
```

```python
for path, save_ in itertools.product(['df_jobs_classified_Warmth_Competence', 'df_jobs_classified_Warmth', 'df_jobs_classified', 'df_jobs_for_analysis'], ['pkl', 'csv']):

    if save_ == 'csv':
        df_jobs = pd.read_csv(f'{df_save_dir}{path}.{save_}', dtype={'Job ID': str})
    else:
        df_jobs = pd.read_pickle(f'{df_save_dir}{path}.{save_}')

    x = pd.get_dummies(df_jobs[['English Requirement in Job Ad', 'Dutch Requirement in Job Ad', 'English Requirement in Sentence', 'Dutch Requirement in Sentence']], dtype='int64')

    df_jobs = df_jobs.drop(columns=['English Requirement in Job Ad_Yes', 'English Requirement in Job Ad_No', 'Dutch Requirement in Job Ad_Yes', 'Dutch Requirement in Job Ad_No', 'English Requirement in Sentence_Yes', 'English Requirement in Sentence_No', 'Dutch Requirement in Sentence_Yes', 'Dutch Requirement in Sentence_No'])

    df_jobs = df_jobs.join(x)

    if ('Warmth' and 'Warmth_Probability') in df_jobs.columns:
        cols_list = cols_list_[:] + ['Warmth', 'Warmth_Probability']
        if ('Competence' and 'Competence_Probability') in df_jobs.columns:
            cols_list = cols_list[:] + ['Competence', 'Competence_Probability']

    df_jobs = df_jobs.loc[:, cols_list]

    print(path.upper(), '.', save_.upper())
    print(df_jobs[['English Requirement in Job Ad_Yes', 'Dutch Requirement in Job Ad_Yes', 'English Requirement in Sentence_Yes', 'Dutch Requirement in Sentence_Yes']].head())

    print(df_jobs[['English Requirement in Job Ad_Yes', 'Dutch Requirement in Job Ad_Yes', 'English Requirement in Sentence_Yes', 'Dutch Requirement in Sentence_Yes']].value_counts())

    print(df_jobs.info())

    if save_ == 'csv':
        df_jobs.to_csv(f'{df_save_dir}{path}.{save_}', index=False)
    else:
        df_jobs.to_pickle(f'{df_save_dir}{path}.{save_}')
```
